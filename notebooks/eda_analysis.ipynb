## lexical co-occurence

For this we have to have soe stopwords that are not to meaningful - mostly the indeclinables

```python
# We'll use our well-prepared 'df_expanded' DataFrame

if 'df_expanded' in locals():
    # 1. Systematically find all unique indeclinable lemmas
    indeclinable_lemmas = set(df_expanded[df_expanded['Word_Type'] == 'Indeclinable']['L'].unique())
    print(f"Found {len(indeclinable_lemmas)} unique indeclinable lemmas to use as stopwords.")
    # e.g., {'ca', 'eva', 'hi', 'tu', 'atha', ...}

    # 2. Add very high-frequency pronouns that act like stopwords
    # While they aren't technically indeclinable, pronouns like 'that', 'I', 'he'
    # often function as stopwords in a network.
    pronoun_stopwords = {'tad', 'mad', 'ka', 'tvad', 'etad', 'yad'}
    print(f"Adding {len(pronoun_stopwords)} high-frequency pronouns to the stopword list.")

    # 3. Combine them into a final stopword set for fast lookup
    combined_stopwords = indeclinable_lemmas.union(pronoun_stopwords)

    print(f"\nTotal number of systematic stopwords: {len(combined_stopwords)}")
    # Showing a small sample of the final list
    print(f"Stopword sample: {list(combined_stopwords)[:10]}")
```

To prepare the data for network analysis, a systematic stopword list was generated by programmatically identifying all indeclinable lemmas from the parsed text. This list was supplemented with high-frequency pronouns that serve a primarily grammatical rather than semantic function. By filtering the text to include only key nominals and adjectives and removing this comprehensive stopword list, we ensure that the resulting lexical network highlights meaningful semantic relationships between core concepts and characters, rather than being cluttered by syntactic noise

```python
if 'df_expanded' in locals() and 'combined_stopwords' in locals():
    # 1. Select the lemmas of important words. As before, we'll select rows
    # where 'Case' is not NaN, as this effectively isolates nouns, adjectives,
    # and some pronouns that we didn't filter.
    all_lemmas = df_expanded[df_expanded['Case'].notna()]['L'].tolist()

    # 2. Filter this list to remove all the words in our new stopword set.
    network_lemmas = [lemma for lemma in all_lemmas if lemma not in combined_stopwords]

    print(f"Original number of important lemmas: {len(all_lemmas)}")
    print(f"Number of lemmas after removing stopwords: {len(network_lemmas)}")
    print("\n--- First 15 lemmas for the final network ---")
    print(network_lemmas[:15])
```

In this step, we will iterate through our list of keywords. We'll use a "sliding window" to decide which words are "near" each other. Every pair of words that appears together inside this window will form a connection (an "edge") in our network. We will also add a 'weight' to each edge, so the more frequently two words appear together, the stronger their connection will be.

```python
import networkx as nx
from itertools import combinations
from collections import defaultdict

if 'network_lemmas' in locals():
    # --- Step 10.1: Create the Co-occurrence Pairs and Weights ---

    # We will store our weighted edges in a dictionary
    # For example: {('prtha', 'duryodhana'): 5} means they appeared together 5 times.
    weighted_edges = defaultdict(int)
    window_size = 7 # Let's use a slightly larger window to capture more context

    # Slide the window across our list of lemmas
    for i in range(len(network_lemmas) - window_size + 1):
        window = network_lemmas[i : i + window_size]
        # Create all unique pairs of lemmas within this window
        for u, v in combinations(window, 2):
            # We sort the pair to ensure ('a', 'b') is treated the same as ('b', 'a')
            edge = tuple(sorted((u, v)))
            weighted_edges[edge] += 1

    # --- Step 10.2: Build the Graph ---

    # Create a new, empty graph object
    G = nx.Graph()

    # Add the nodes and weighted edges from our dictionary
    for edge, weight in weighted_edges.items():
        # The 'weight' attribute is now stored with each edge
        G.add_edge(edge[0], edge[1], weight=weight)

    print("Network built successfully!")
    print(f"The graph has {G.number_of_nodes()} nodes (unique words).")
    print(f"The graph has {G.number_of_edges()} edges (unique connections).")
```

```python
node_data = []
for node, attributes in G.nodes(data=True):
    node_data.append({"Node": node, **attributes})

nodes_df = pd.DataFrame(node_data)
```

We will draw a static image of the graph. To make it professional and easy to read, the size and color of each word (node) will be based on its importance in the network. We'll measure importance using degree centrality, which is simply the number of connections a node has.

```python
import matplotlib.pyplot as plt

if 'G' in locals() and G.number_of_nodes() > 0:
    plt.figure(figsize=(20, 20)) # A large figure size is key for readability

    # 1. Use a spring layout algorithm to position nodes. `k` adjusts spacing.
    # A random `seed` makes the layout reproducible every time you run it.
    pos = nx.spring_layout(G, k=0.1, iterations=50, seed=42)

    # 2. Calculate node degrees to determine size and color
    degrees = dict(G.degree())
    node_sizes = [v * 100 for v in degrees.values()] # Scale size
    node_colors = [v for v in degrees.values()] # Color by degree

    # 3. Draw the nodes and labels
    nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors,
                           cmap=plt.cm.plasma, alpha=0.9)
    nx.draw_networkx_labels(G, pos, font_size=12, font_color='black')

    # 4. Draw the edges, with transparency to de-clutter
    # Let's make the edge width proportional to its weight for more insight
    edge_weights = [G[u][v]['weight'] * 0.25 for u, v in G.edges()]
    nx.draw_networkx_edges(G, pos, width=edge_weights, alpha=0.2, edge_color='grey')

    # 5. Final customization
    plt.title('Lexical Co-occurrence Network of the Sanskrit Text', fontsize=24)
    plt.axis('off')
    plt.show()
```

```python
# ==============================================
# PMI: adjectives most associated with characters
# ==============================================
from math import log2
from collections import defaultdict

# Ensure we have sentence docs
try:
    sent_docs
except NameError:
    doc = nlp(text)
    sent_docs = list(doc.sents)

def norm(s: str):
    import unicodedata, re
    s = unicodedata.normalize("NFD", s)
    s = "".join(ch for ch in s if unicodedata.category(ch) != "Mn")
    return s.lower()

# Build sentence-level sets of (characters present) and (adjectives present)
N = 0
char_in_sent = defaultdict(int)           # count of sentences mentioning char
adj_in_sent = defaultdict(int)            # count of sentences containing adjective
joint_in_sent = defaultdict(lambda: defaultdict(int))  # char -> adj -> joint count

for sent in sent_docs:
    s_text = sent.text
    N += 1

    # which characters (via aliases) occur in this sentence?
    n_sent = norm(s_text)
    chars_here = set()
    for a_norm, canon in alias_to_canonical.items():
        if canon in canonical_chars and re.search(rf"(^|[^a-z]){re.escape(a_norm)}($|[^a-z])", n_sent):
            chars_here.add(canon)

    # adjectives in this sentence (lemma, lowercase), unique per sentence
    adjs_here = set(t.lemma_.lower() for t in sent if t.pos_ == "ADJ")

    for c in chars_here:
        char_in_sent[c] += 1
    for a in adjs_here:
        adj_in_sent[a] += 1
    for c in chars_here:
        for a in adjs_here:
            joint_in_sent[c][a] += 1

# Compute PMI and show top collocates per character
min_joint = 1   # raise to 2 if you want to be stricter
top_k = 8

print("\nTop PMI adjective collocates per character:")
for c in canonical_chars:
    c_count = char_in_sent.get(c, 0)
    if c_count == 0:
        continue
    rows = []
    for a, j in joint_in_sent[c].items():
        a_count = adj_in_sent.get(a, 0)
        if j >= min_joint and a_count > 0:
            # PMI = log2( (j/N) / ((c_count/N)*(a_count/N)) ) = log2(j*N/(c_count*a_count))
            pmi = log2(j * N / (c_count * a_count))
            rows.append((a, j, a_count, pmi))
    if not rows:
        continue
    # sort by PMI then by joint frequency
    rows.sort(key=lambda x: (x[3], x[1]), reverse=True)
    pretty = ", ".join([f"{a} (PMI {pmi:.2f}, n={j})" for a, j, _, pmi in rows[:top_k]])
    print(f"- {c}: {pretty}")

# Optional: export a tidy DataFrame for later analysis
try:
    import pandas as pd
    recs = []
    for c in canonical_chars:
        for a, j in joint_in_sent[c].items():
            if j >= min_joint and adj_in_sent.get(a,0) > 0 and char_in_sent.get(c,0) > 0:
                pmi = log2(j * N / (char_in_sent[c] * adj_in_sent[a]))
                recs.append({'character': c, 'adjective': a, 'joint': j,
                             'char_sent': char_in_sent[c], 'adj_sent': adj_in_sent[a], 'PMI': pmi})
    pmi_df = pd.DataFrame(recs).sort_values(['character','PMI','joint'], ascending=[True, False, False])
    if not pmi_df.empty:
        print("\nPMI table head:")
        print(pmi_df.head(15).to_string(index=False, float_format=lambda x: f"{x:.2f}"))
except Exception:
    pass
```

```python
# --- Start of Corrected Code Block ---

# First, check if the graph is connected and find its components
if not nx.is_connected(G):
    print(f"The graph is not connected. It has {nx.number_connected_components(G)} separate components.")

    # Find the largest connected component
    largest_cc = max(nx.connected_components(G), key=len)

    # Create a new graph (subgraph) containing only the largest component
    G_main = G.subgraph(largest_cc)

    print(f"Proceeding with analysis on the largest component, which has {G_main.number_of_nodes()} nodes.")

    # Calculate centrality measures on this main component
    degree_centrality = nx.degree_centrality(G_main)
    betweenness_centrality = nx.betweenness_centrality(G_main)
    # The eigenvector centrality will now converge on the connected subgraph
    eigenvector_centrality = nx.eigenvector_centrality(G_main)

else:
    print("The graph is connected. Running analysis on the full graph.")
    G_main = G # Use the original graph if it's already connected
    degree_centrality = nx.degree_centrality(G_main)
    betweenness_centrality = nx.betweenness_centrality(G_main)
    eigenvector_centrality = nx.eigenvector_centrality(G_main)


# Create a DataFrame for centrality measures from the main component
centrality_df = pd.DataFrame({
    'Character': list(degree_centrality.keys()),
    'Degree Centrality': list(degree_centrality.values()),
    'Betweenness Centrality': list(betweenness_centrality.values()),
    'Eigenvector Centrality': list(eigenvector_centrality.values())
})

# Sort by Degree Centrality to see the most connected characters
centrality_df = centrality_df.sort_values(by="Degree Centrality", ascending=False)

print("\nNetwork Centrality Measures (for the largest connected component):")
print(centrality_df)

# Plotting the centrality measures for better visualization
plt.figure(figsize=(12, 6))
sns.barplot(data=centrality_df, x='Character', y='Degree Centrality', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Degree Centrality of Characters')
plt.xlabel('Character')
plt.ylabel('Degree Centrality')
plt.tight_layout()
plt.show()

# --- End of Corrected Code Block ---
```
